## Data Cleaning and Preprocessing

Before starting the analysis, it is crucial to clean and prepare the raw data. This step involves addressing several key issues observed in the initial dataset:
- **Handling Missing Values:** Dropping rows with missing CustomerID.
- **Filtering Invalid Data:** Removing transactions with zero or negative quantities and prices.
- **Data Type Conversion:** Converting the 'InvoiceDate' column to a datetime format for time-based analysis.
- **Feature Engineering:** Creating a 'TotalPrice' column to calculate total sales for each transaction.

```python
# Step 2: Data Preprocessing and Cleaning

# We'll use the file name from the latest upload.
df = pd.read_csv('data (2).csv', encoding='latin-1')

# Let's first inspect the initial state of the data
print("Initial Data Information:")
df.info()
print("\nInitial Descriptive Statistics:")
print(df.describe())

# 2.1 Handling Missing Values
# Drop rows where CustomerID is missing, as they are crucial for our analysis.
df.dropna(subset=['CustomerID'], inplace=True)

# 2.2 Filtering Out Invalid Data
# Remove rows with a Quantity <= 0 or UnitPrice <= 0
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# 2.3 Converting Data Types
# Convert InvoiceDate from object to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# 2.4 Feature Engineering
# Create a new column 'TotalPrice'
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Display the information and statistics of the cleaned data
print("\n--- Data Cleaning Complete ---")
print("Cleaned Data Information:")
df.info()
print("\nCleaned Descriptive Statistics:")
print(df.describe())
```

## Analysis and Key Findings

Following the data cleaning process, we performed several analyses to gain a better understanding of the business performance.

### Monthly Sales Trend

The first analysis focuses on the company's monthly sales performance to identify seasonal patterns and overall growth. The following chart visualizes the total sales over time.

```python
# Monthly sales trend analysis
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
monthly_sales = df.groupby('InvoiceMonth')['TotalPrice'].sum()

# Visualize the monthly sales trend
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o', color='purple')
plt.title('Monthly Total Sales Trend', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales (GBP)', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('monthly_sales_trend.png')

print("Monthly sales trend chart has been created and saved as 'monthly_sales_trend.png'.")
```

### Top 10 Best-Selling Products

Understanding which products are most popular is vital for inventory management and marketing strategies. This analysis identifies the top 10 products by total quantity sold.

```python
# Step 3: Analysis for Business Insights

# 3.1 Top 10 Best-Selling Products by Quantity

# Group the data by 'Description' and sum the 'Quantity' for each product.
# Then, sort the values in descending order and get the top 10.
top_selling_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

print("Top 10 Best-Selling Products:")
print(top_selling_products)

# Visualize the top-selling products
plt.figure(figsize=(12, 6))
top_selling_products.plot(kind='barh', color='skyblue')
plt.title('Top 10 Best-Selling Products by Quantity', fontsize=16)
plt.xlabel('Total Quantity Sold', fontsize=12)
plt.ylabel('Product Description', fontsize=12)
plt.gca().invert_yaxis()  # Put the best-selling product on top
plt.tight_layout()
plt.savefig('top_selling_products.png')

print("\n--- The bar chart of top-selling products has been created and saved as 'top_selling_products.png'. ---")
```

### Top 10 Customers by Spending

Identifying high-value customers is essential for customer relationship management (CRM). This section highlights the top 10 customers based on their total spending.

```python
# Step 3.2: Identifying Top 10 Customers by Total Spending

# Group the data by 'CustomerID' and sum the 'TotalPrice' for each customer.
# Then, sort the values in descending order and get the top 10.
top_customers = df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).head(10)

print("\nTop 10 Customers by Total Spending:")
print(top_customers)

# Visualize the top customers
plt.figure(figsize=(12, 6))
top_customers.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Customers by Total Spending', fontsize=16)
plt.xlabel('Customer ID', fontsize=12)
plt.ylabel('Total Spending (GBP)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_customers.png')

print("\n--- The bar chart of top customers has been created and saved as 'top_customers.png'. ---")
```
